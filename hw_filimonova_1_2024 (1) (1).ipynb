### Задача 1
Изобразите график параметрически заданной функции. Переменные необходимо задавать с помощью Numpy.

$$ x=\alpha\times\left(\cos t+\frac{cos(\beta t)}{\beta}\right)$$

$$y=\alpha\times\left(\sin t−\frac{sin(\beta t)}{\beta}\right)$$

$$t∈[0;20π]$$

1. Реализуйте функции $x(t, \alpha, \beta)$ и $y(t, \alpha, \beta)$

```python
t=np.linspace(0,21*np.pi)

def X(t,alfa,beta):
  return alfa*(np.cos(t)+np.cos(beta*t)/beta)

def Y(t,alfa,beta):
  return alfa*(np.sin(t)-np.sin(beta*t)/beta)
```

2. Постройте график в осях $x$ и $y$ при $\alpha=4$ и $\beta=1.1$. Подпишите
оси.

```python
#зададим параметры
alfa= 4
beta=1.1

plt.figure(figsize=(15,10))
plt.plot(X(t,alfa,beta), Y(t,alfa,beta))
plt.grid()

plt.title('График функции (x(t,' r'$\alpha$' ',' r'$\beta$' ' ), y(t,' r'$\alpha$' ', ' r'$\beta$' ' )) при  ' r'$\alpha=4$' ',' r'$\beta$=1.1' ' ' , fontsize=20)
plt.xlabel('$x$', fontsize=20)
plt.ylabel('$y$',fontsize=20)
plt.show()
```

#### Ядерные оценки плотности

Иногда гистограммы недостаточно, чтобы понять, какое у распределение у элементов выборки. В таких случаях можно использовать ядерные оценки плотности.

В библиотеке `seaborn` они представлены функциями `distplot` (в версиях до `0.11.1`), `kdeplot` (начиная с версии `0.11.1`)

Пример для нормального распределения с выбросом.

```python
plt.figure(figsize=(15, 6))
sns.distplot(sample, bins=20);
```

Видно, что, даже несмотря на "хвост", плотность основной части данных довольно похожа на плотность нормального распределения. Теперь нарисуем этот же график для выборки без выброса.

Для наглядности на этом же графике нарисуем плотность стандартного нормального распределения. В библиотеке `scipy` плотность распределения называется `pdf`.

```python
# уберем выброс
sample_wo_ourliers = sample[1:]

# сетка от -4 до 4 из 1000 значений
grid = np.linspace(-4, 4, 1000)
# считаем значения функции в выборке
norm_pdf_values = sps.norm.pdf(grid)

# отрисовка графика
plt.figure(figsize=(15, 6))
sns.distplot(sample_wo_ourliers, bins=20, kde_kws={"label": "KDE", "linewidth": 2})
plt.plot(grid, norm_pdf_values, label='Плотность $\\mathcal{N}(0,1)$', lw=2)
plt.title('Сравнение оценки плотности с истинной')
plt.legend();
```

Однако ядерные оценки плотности тоже не универсальны.
Рассмотрим пример экспоненциального распределения.

Сгенерируйте выборку размера 100 из экспоненциального распределения $\textit{Exp(1)}$. Для этого воспользуйтесь функцией `sps.expon.rvs`.

```python
samples_exp=sps.expon(scale=1).rvs(size=100)
```

Посчитайте значения плотности этого распределения на промежутке от 0 до 6.

```python
#сетка от 0 до 6 из 1000 значений
grid_exp=np.linspace(0,6,1000)
# считаем значения плотности распределения
pdf_exp=sps.expon.pdf(grid_exp)
# выведем полученные значения для наглядности
print(pdf_exp)
```

Нарисуйте гистограмму и ядерную оценку плотности для выборки из сгененированного экспоненциального распределения.

```python
#отрисуем гистограмму и KDE
plt.figure(figsize=(15,6))
sns.distplot(samples_exp, bins=20,kde_kws={"label":"KDE", "linewidth" :2 })

#отрисуем плотность распределения
plt.plot(grid_exp, pdf_exp, label='Плотность $\\mathcal {Exp}(1)$', lw=2)
plt.title('Гистограмма экспоненциального распределения', fontsize=20)
plt.legend()

plt.show()
```

В этой задаче нужно визуализировать *центральную предельную теорему*.



*a).* Пусть $\xi_1, ..., \xi_n$ — независимые случайные величины из распределения $Exp(\lambda)$. Согласно центральной предельной теореме выполнена сходимость $Z_n = \frac{X_n - \mathsf{E}X_n}{\sqrt{\mathsf{D}X_n}} \stackrel{d}{\to} \mathcal{N}(0, 1)$, где $X_n = \sum\limits_{i=1}^n \xi_i$. Вам нужно убедиться в этом, сгенерировав множество наборов случайных величин и посчитав по каждому из наборов величину $Z_n$ в зависимости от размера набора.

Сгенерируйте 500 наборов случайных величин $\xi_1^j, ..., \xi_{300}^j$ из распределения $Exp(1)$.

```python
size = 300  # размер выборки
samples_count = 500  # количество выборок

samples = sps.expon(scale=1).rvs(size = (samples_count, size))
```

По каждому из них посчитайте сумму $X_{jn} = \sum\limits_{i=1}^n \xi_i^j$ для $1 \leqslant n \leqslant 300$, то есть сумма первых $n$ величин $j$-го набора.

```python
# посчитаем по каждому из них сумму
def summ_j(j):
  summ = samples[j].cumsum() # по каждому набору проведем суммирование методом частичных сумм
  return summ
```

Для этого среднего посчитайте величину $Z_{jn} = \frac{X_{jn} - \mathsf{E}X_{jn}}{\sqrt{\mathsf{D}X_{jn}}}$.

*Подсказка: можно воспользоваться возможностями библиотеки `numpy`.*

```python
def Z_jn(j):

  z = np.ones(300)
  for n in range(0, 300, 1):
    z[n] = (summ_j(j)[n] - (n + 1))/np.sqrt(n + 1)
  return z
```

Для каждого $j$ нанесите на один график зависимость $Z_{jn}$ от $n$ с помощью `plt.plot`. Каждая кривая должна быть нарисована *одним цветом* с прозрачностью `alpha=0.05`. Сходятся ли значения $Z_{jn}$ к какой-либо константе?

```python
plt.figure(figsize=(15,7))

#для каждой выборки построим отдельно график
for j  in range (samples_count):
  x=np.linspace(1,301,300)
  plt.plot(x, Z_jn(j),color='cyan',alpha=0.05)


plt.xlabel('$Z_{jn}$')
plt.ylabel('Количество случайных величин')
plt.xlim((0,size))
plt.show()
```


Для $n=300$ по набору случайных величин $Z_{1,300}, ..., Z_{500,300}$ постройте гистограмму. Похожа ли она на плотность распределения $\mathcal{N}(0, 1)$, которую тоже постройте на том же графике? Не забудьте сделать легенду (подписи на графиках, см. семинар и туториалы).

```python
plt.figure(figsize=(15,6))

#создадим массив , количество элементов которого совпадает с количеством выборок
y=np.array([Z_jn(j)[299] for j in range(samples_count)])

sns.histplot(y, bins=20,kde=True,label='KDE', lw=2 ,stat ='density')

# по гистограмме значения меняются от -3 до 3
grid=np.linspace(-3,3,1000)
pdf_Norm=sps.norm.pdf(grid)   # плотность нормального распределения
plt.plot(grid, pdf_Norm, color='red', label='Плотность $\\mathcal{N}(0,1)$', lw=2)
plt.legend()
```

Полученная гистограмма похожа на плотность нормального распределения.

```python
size = 300  #размер и количество выборот оставим теми же
samples_count = 500

samples = sps.poisson(mu=1).rvs(size = (samples_count, size))
samples
```

б)По каждому из них посчитайте сумму $X_{jn} = \sum\limits_{i=1}^n \xi_i^j$ для $1 \leqslant n \leqslant 300$

```python
# функцию просто перенесли, чтобы все было в одном месте
def summ2_j(j):
  summ2 = samples[j].cumsum()
  return summ2
```

в)Для этого среднего посчитайте величину $Z_{jn} = \frac{X_{jn} - \mathsf{E}X_{jn}}{\sqrt{\mathsf{D}X_{jn}}}$.
$Poisson(1)$, $\lambda = 1$, значит $EX = \dfrac{1}{\lambda} = 1$, $DX = \dfrac{1}{\lambda ^2} = 1$.

```python
def Z2_jn(j):
  z2 = np.ones(300)
  for n in range(0, 300, 1):
    z2[n] = (summ2_j(j)[n] - (n + 1))/np.sqrt(n + 1)
  return z2
```

Для каждого $j$ нанесите на один график зависимость $Z_{jn}$ от $n$

```python
plt.figure(figsize=(15,7))

for j  in range (samples_count):
  x=np.linspace(1,301,300)
  plt.plot(x, Z2_jn(j),color='green',alpha=0.05)


plt.xlabel('$Z_{jn}$')
plt.ylabel('Количество случайных величин')
plt.xlim((0,size))
plt.show()
```


Для $n=300$ по набору случайных величин $Z_{1,300}, ..., Z_{500,300}$ постройте гистограмму. Также построим на графике плотность нормального рапределения, сравним гистограмму и график.

```python
plt.figure(figsize=(15,6))

#создадим массив , количество элементов которого совпадает с количеством выборок
y=np.array([Z2_jn(j)[299] for j in range(samples_count)])

sns.histplot(y, bins=20,kde=True,label='KDE', lw=2 ,stat ='density')

# по гистограмме значения меняются от -3 до 3
grid=np.linspace(-3,3,1000)
pdf_Norm=sps.norm.pdf(grid)   # плотность нормального распределения
plt.plot(grid, pdf_Norm, color='red', label='Плотность $\\mathcal{N}(0,1)', lw=2)
plt.legend()
```

**Сделаем вывод:** сравнив гистограммы и плотность нормального распределения, можно сказать,что распределения случайных величин(и экспоненциальное, и пуассоновское) стремятся к нормальному.
